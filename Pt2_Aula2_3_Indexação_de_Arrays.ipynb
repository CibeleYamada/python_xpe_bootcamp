#**INDEXAÇÃO DE ARRAYS**

- acessando elementos do array.


Relembrando:

- o numpy array é uma estruturas de números ordenados; esses números ocupam índices.
- os índices de um elementos iniciam a partir do 0

image.png



```python
import numpy as np
```

**Acessando elementos de um array 1D**

- sintaxe: 

nome_do_array[posição do elemento]

```python
# Acessando elementos de uma matriz 1D: dada a matriz acima, queremos acessar o elemento da posição 1:

A = np.array([88,19,46,74,94])      # criando a matriz

A[1]    # acessando o elemento da posição 1
```

```python
# Acessando o último elemento do array: posição -1

A[-1]
```

**Acessando elementos de uma matriz 2D**

- sintaxe: devemos passar a posição do eixo x (linha) e do eixo y (coluna). Lembrando que eles também iniciam em 0.

nomedamatriz[posição_x, posição_y]

```python
# Criando o array 2D

B = np.array([[88,19,46,74,94],[69,79,26,7,29],[21,45,12,80,72]])
print(B)
```

```python
# Acessando o elemento 45, que está na linha 2 e coluna 1:

B[2,1]
```

#**Slicing (= fatiamento)**

- acesso a mais de um elemento em um array (acesso de um intervalo).
- sintaxe: entenda os dois pontos (:) como se fosse "até"

nomedoarray[posição de inicio **:** uma posição a mais da ultima posição que você quer]

```python
# Vamos acessar os elementos 19 e 46 do array A, cujas posições são 1 e 2.
# lembrando que devemos inserir 1 posição a mais da última posição que quereemos, então, o intervalo fica 1:3

A[1:3]
```

```python
# Quando omitimos a posição do 1º elemento, ele considera o 0:

print(A[:3])

# significa que estamos acessando os elementos entre 0 a 2
```

```python
# Também podemos omitir a posição do último elemento, que ele considerará -1
# a contagem das posições será invertida. Ex: acessando os dois últimos elementos:

A[-2:]
```

```python
# Quando omitimos a posição do 1º e o 2º que queremos encontrar, ele retorna todo o array.

A[:]
```

OBS: a operação de SLICING retorna um numpy array reduzido.

**Slicing em array 2D**

```python
B[1:3, 1:4]
```

**OBS**: quando fazemos um slice, geramos um numpy array reduzido de um array original. Se *alterarmos* o valor de um dos elementos do array reduzido, automaticamente, o valor desse elemento no array original também será ALTERADO.

- isso porque esse array reduzido (subarray) compartilha o mesmo espaço da memória do array original

```python
# Ex: array original

print(A)
```

```python
# vamos pegar um slice do array original

a = A[0:3]
print(a)
```

```python
# vamos alterar o valor do elemento 46 para 32:

a[2] = 32

# imprimindo o array original A após a alteração do valor no subarray:

print(A)
```

Caso você não queira que isso ocorra, podemos utilizar o método **.copy** no momento em que você cria o slice.
- nesse caso, estamos gerando uma cópia do array original e qualquer modificação neste subarray modificado, não alterará o array original.

```python
print(A)

# lembrando que modificamos o array original, então consideraremos esse nosso novo array original
```

```python
b = A[0:3].copy()      # gerando um subarray como cópia

print(b)
```

```python
b[2] = 100      # alterando o valor do elemento da posição 3: de 32 para 100

print(b)
```

```python
# imprimindo o array original:

print(A)
```
