```python
import numpy as np
```

```python
# Antes de tudo, devemos importar a biblioteca matplotlib, que é uma biblioteca para visualização de dados.


import matplotlib.pyplot as plt
```

```python
# dados:

x = [-1., -0.77777778, -0.55555556, -0.33333333, -0.11111111,
     0.11111111, 0.33333333, 0.55555556, 0.77777778, 1.]

y = [-1.13956201, -0.57177999, -0.21697033, 0.5425699, 0.4940665,
     1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.95684202]
```

```python
# Agora vamos plotar os dados:
# 1º passo é instanciar um objeto de figura; nesse caso, estamos passando o tamanho da figura como argumento.

plt.figure(figsize=(10, 5))       
plt.plot(x, y, 'o', label='dados originais')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

**Regressão linear**

Dada uma matriz x, em que o número de linhas será a quantidade de amostras coletadas e o número de colunas será o número de atributos que ela possui, temos:




regressão linear.png

```python
"""
Iremos estimar uma função do tipo: y = a*x + b
ou seja, devemos achar quais os valores de a e b
que melhor representem os dados.

"""

# Primeiro, note que estamos trabalhando com x e y sendo listas. Devemos transformá-los em arrays.
# Transformando x e y para numpy e queremos que ele tenha 1 coluna (reshape):
# Obs: no método reshape, quando informamos que um elemento é -1 em uma determinada dimensão (que nesse caso, é a linha), queremos informar que estamos ignorando o número de linhas. Nesse exemplo, estamos forçando a matriz a ter 1 coluna, e ele vai calcular automaticamente o número de linhas, já que informamos que não nos importa (-1). 

x, y = np.array(x).reshape(-1,1), np.array(y).reshape(-1,1)

# O próximo passo é adicionar uma coluna de 1 no array x; o np.hstack faz a concatenação de dois arrays horizontalmente.

X = np.hstack((x, np.ones(x.shape)))

# Agora, para estimar a e b:

beta = np.linalg.pinv(X).dot(y)
print('a estimado:', beta[0][0])
print('b estimado:', beta[1][0])
```

```python
#plot dos dados:

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, X.dot(beta), label='regressão linear')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão linear com numpy')
plt.grid()
plt.show()
```
