- Inicialmente precisamos importar a biblioteca Pandas:

```python
import pandas as pd
```

**Leitura de dados**

```python
# leitura de dados CSV:

df = pd.read_csv('https://pycourse.s3.amazonaws.com/temperature.csv')
df
```

```python
# leitura de planilha de excel (arquivo todo)
# obs: o arquivo tem 2 planilhas

excel_file = pd.ExcelFile('https://pycourse.s3.amazonaws.com/temperature.xlsx')
```

```python
# leitura da primeira planilha do excel (com nome Sheet1 - é o nome padrão que o excel dá)
# Arquivo com dados numéricos com separador decimal = '.' (que é o default (padrão)).

df2 = pd.read_excel(excel_file, sheet_name='Sheet1')
df2
```

```python
# vamos fazer a leitura da 2ª planilha do excel (Sheet2), usando como separador decimal a ',':

df3 = pd.read_excel(excel_file, sheet_name='Sheet2', decimal=',')
df3
```

```python
# vamos fazer a leitura da 2ª planilha, mas sem especificar o separador decimal:

df4 = pd.read_excel(excel_file, sheet_name='Sheet2')
df4

# obs: note que, na coluna 'temperatura', os dados foram importados com a vírgula (que é o separador usado na tabela original); porém, ao importá-los pro Python, não conseguimos usar esses dados, a não ser que tranformemos a vírgula em ponto.
# desse modo, esses dados com vírgula são tratados pelo Pandas como dados do tipo OBJETO, e não, FLOAT.
```
