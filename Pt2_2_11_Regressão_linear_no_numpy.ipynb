#**REGRESSÃO LINEAR NO NUMPY**

Dado um conjunto de pontos, queremos achar qual a função que melhor descreve esses pontos.

image.png

E como podemos fazer isso? Existem várias formas. Podemos, por exemplo, traçar uma reta entre os dois pontos extremos. Mas será que esta é uma estratégia robusta suficiente? Nem semprea ligação entre dois pontos representa o que realmente queremos. 

É por isso, que devemos escolher a técnica baseada em alguns critérios, como:

**Erro quadrático**: o erro é basicamente a diferença entre o valor real do dado e o valor que função 'diz' que ele tem. Quadrático, porque não interessa se estamos errando pra mais ou pra menos.

- com o critério definido e, sabenddo que a função linear possui o seguinte formato: **y = f(x) = ax + b**, o problema de regressão linear resume-se à determinação dos coeficientes **a** e **b**, visto que **x** e **y** são dados de entrada.
